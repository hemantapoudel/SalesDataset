```python
import pandas as pd
```

```python
# Load the sales data into a pandas DataFrame
df = pd.read_csv('sales_data.csv', delimiter=';')
```

```python
# Display the first 5 rows of the dataset
print(df.head())
```

```python
# Check for any missing values in the dataset
print(df.isnull().sum())
```

```python
# Add a new column TotalSales that contains the total sales for each product of an order (i.e., QuantityOrdered * PriceEach).
df['QuantityOrdered'] = pd.to_numeric(df['QuantityOrdered'], errors='coerce')
df['PriceEach'] = pd.to_numeric(df['PriceEach'], errors='coerce')
df['TotalSales'] = df['QuantityOrdered'] * df['PriceEach']
print(df.columns)
```

```python
#Find the total sales across all orders.
total_sales = df['TotalSales'].sum()
print("Total Sales across all orders:", total_sales)
```

```python
# Identify the product that generated the highest total sales (across all orders).
# Group by ProductCode and sum the TotalSales for each product
product_sales = df.groupby('ProductCode')['TotalSales'].sum()

# Find the product with the highest total sales
highest_sales_product = product_sales.idxmax()
highest_sales_value = product_sales.max()

print("Product with the highest total sales:", highest_sales_product)
print("Highest total sales value:", highest_sales_value)
```

```python
# Calculate the average sales per order
average_sales_per_order = df['TotalSales'].mean()
print("Average sales per order:", average_sales_per_order)
```

```python
#Determine the city with the highest number of orders.
city = df.groupby('City')['QuantityOrdered'].sum()
city_max_order = city.idxmax()
print("City with the highest number of orders:", city_max_order)
```

```python
city_orders = df[df['City'] == city_max_order]

# Count the unique OrderIDs for Madrid
unique_orders = city_orders['OrderID'].nunique()

print(f"Number of unique orders from {city_max_order}: {unique_orders}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by City and sum the TotalSales for each city
city_sales = df.groupby('City')['TotalSales'].sum().sort_values(ascending=False)

# Extract cities and sales values for plotting
cities = city_sales.index
sales = city_sales.values

# Plotting the bar chart with plt.bar
plt.figure(figsize=(15, 6))
plt.bar(cities, sales, color='teal')

# Adding labels and title
plt.xlabel('City')
plt.ylabel('Total Sales (Thousands of Dollars)')
plt.title('Total Sales per City')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
